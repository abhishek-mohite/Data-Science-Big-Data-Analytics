```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic_data = sns.load_dataset('titanic')

# Set up the figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot histogram with KDE
sns.histplot(titanic_data['fare'], kde=True, ax=axes[0])
axes[0].set_title('Fare Distribution with KDE')
axes[0].set_xlabel('Fare')
axes[0].set_ylabel('Frequency')

# Plot histogram without KDE
sns.histplot(titanic_data['fare'], kde=False, ax=axes[1])
axes[1].set_title('Fare Distribution without KDE')
axes[1].set_xlabel('Fare')
axes[1].set_ylabel('Frequency')

# Show plots
plt.tight_layout()
plt.show()
```
