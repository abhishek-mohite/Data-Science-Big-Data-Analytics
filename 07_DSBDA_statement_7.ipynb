```python
import pandas as pd
import numpy as np
df=pd.read_csv('iris.csv');
df
```

```python
# 1.	Standard Scaler and Minimax Scaler operation on Iris Data set.
from sklearn.preprocessing import StandardScaler
# caeate instance
scaler=StandardScaler()
col_names=["sepal.length","sepal.width",'petal.length','petal.width']
features=df[col_names];
# print(features)
model=scaler.fit_transform(features);
model=pd.DataFrame(data=model,columns=col_names)
model["variety"]=df['variety']
print(model.head())
# value present in between -2 t0 2
```

```python
# 1.	Standard Scaler and Minimax Scaler operation on Iris Data set.
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
col_names=["sepal.length","sepal.width",'petal.length','petal.width']
features=df[col_names];
# print(features)
model=scaler.fit_transform(features);
model=pd.DataFrame(data=model,columns=col_names)
model["variety"]=df['variety']
model
```

```python
# 2.	Scale data with range 5 to 10 using Minimax Scaler operation on Iris Data set.
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(5,10))
col_names=["sepal.length","sepal.width",'petal.length','petal.width']
features=df[col_names];
# print(features)
model=scaler.fit_transform(features);
model=pd.DataFrame(data=model,columns=col_names)
model["variety"]=df['variety']
model
```

```python
# 3.	Write a Python code for outlier detection using Z score
threshold=3
data=df["sepal.length"]
mean=np.mean(data)
std=np.std(data)
print("mean",mean)
print("std",std)
outlier=[]
for i in data:
    z=(i-mean)/std
    if z>threshold:
        outlier.append(i)

print(outlier)
```
